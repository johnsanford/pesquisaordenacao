```
import random

def createList(size):
    lyst = []
    for x in range(size):
        lyst.append(random.randrange(100000))
    return lyst

unsortedList = createList(1000)
print(unsortedList)

def createList2(size):
    lyst = []
    for x in range(size):
        lyst.append(random.randrange(100000))
    return lyst

unsortedList2 = createList2(10000)
print(unsortedList2)

def createList3(size):
    lyst = []
    for x in range(size):
        lyst.append(random.randrange(100000))
    return lyst

unsortedList3 = createList3(30000)
print(unsortedList3)

def createList4(size):
    lyst = []
    for x in range(size):
        lyst.append(random.randrange(100000))
    return lyst

unsortedList4 = createList4(60000)
print(unsortedList4)

from random import randint
import timeit
import matplotlib as mpl
import matplotlib.pyplot as plt


def mergeSort(lyst):

   print("Separando ",lyst)

   if len(lyst)>1:
       mid = len(lyst)//2
       lefthalf = lyst[:mid]
       righthalf = lyst[mid:]

       #recursion
       mergeSort(lefthalf)
       mergeSort(righthalf)

       i=0
       j=0
       k=0

       while i < len(lefthalf) and j < len(righthalf):
           if lefthalf[i] < righthalf[j]:
               lyst[k]=lefthalf[i]
               i=i+1
           else:
               lyst[k]=righthalf[j]
               j=j+1
           k=k+1

       while i < len(lefthalf):
           lyst[k]=lefthalf[i]
           i=i+1
           k=k+1

       while j < len(righthalf):
           lyst[k]=righthalf[j]
           j=j+1
           k=k+1

   print("Mesclando ",lyst)

mergeSort(unsortedList)
mergeSort(unsortedList2)
mergeSort(unsortedList3)
mergeSort(unsortedList4)
print(unsortedList)
print(unsortedList2)
print(unsortedList3)
print(unsortedList4)

mpl.use('Agg')
  
def desenhaGrafico(x,y,y2,xl = "Entradas", yl = "SaÃ­das"):
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111)
    ax.plot(x,y, label = "1000")
    ax.plot(x,y2, label = "10000")
    ax.plot(x,y3, label = "30000")
    ax.plot(x,y4, label = "60000")
    ax.legend(bbox_to_anchor=(1, 1),bbox_transform=plt.gcf().transFigure)
    plt.ylabel(yl)
    plt.xlabel(xl)
    fig.savefig('MergeSort.png')
   
x2 = [1,2,3,4]
y = []
y2= []
y3= []
y4= []
  
for i in range(4):
    y.append(timeit.timeit("createList({})".format(x2[i]),setup="from __main__ import createList",number=1))  
    y2.append(timeit.timeit("createList2({})".format(x2[i]),setup="from __main__ import createList2",number=1)) 
    y3.append(timeit.timeit("createList3({})".format(x2[i]),setup="from __main__ import createList3",number=1)) 
    y4.append(timeit.timeit("createList4({})".format(x2[i]),setup="from __main__ import createList4",number=1)) 
  
desenhaGrafico(x2,y,y2)
```
